I think that deep neural networks have already proven themselves to be a major breakthrough. While they likely still have limitations that we're yet to discover, the advancements that deep neural networks have allowed us to make turn them into a significant advancement in AI history no matter what their end limitation ends up being. Areas such as AI text generation, image classification, self-driving cars, and language translation have all been so significantly advanced by the era of deep neural networks that this technology has already made its mark as a significant step forward.

Do deep neural networks have limitations? I think undoubtedly. Even Watson, the deep-neural-net-trained AI that bested Jeopardy's human champions still gave completely off the wall, wildly incorrect answers on occasion. But then again, Watson was an upgrade of Deep Blue, which beat humanity's chess grandmasters, and since Watson AI has progressed even further with AlphaGo and AlphaGo Zero. Perhaps deep neural networks have no limitations in theory, but in practice the neural network size and the amount of data and training required to acquire "perfect" or even close to perfect performance might be too high to be practical for some of the applications of AI that still only exist in science fiction. While specialized performance in a single, specialized area has been shown to be very accessible through deep neural nets (making AI Jeopardy or chess champions), deep neural networks haven't even yet managed to create an AI that can play a sport as well as a human. Self-driving cars aren't even widespread yet, and the motor skills and decision making required to drive a car around a city are significantly less intensive than those employed by the world's top athletes.

But overall, I think deep neural nets are a fantastic step in the right direction. However, significantly greater advancements will have to be made before we have our first AI Olympian, and even more steps will have to be taken before we see anything like Jarvis from Iron Man, or (heaven forbid) Skynet from the Terminator franchise.

```python
from tensorflow.keras import layers
from tensorflow.keras import models
```

```python
model = models.Sequential()

# Configure a convnet with 3 layers of convolutions and max pooling.
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (4, 4), activation='relu'))

# Add layers to flatten the 2D image and then do a 10-way classification.
model.add(layers.Flatten())
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

model.summary()
```

```python
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
train_images = train_images.reshape((60000, 28, 28, 1))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28, 28, 1))
test_images = test_images.astype('float32') / 255

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(train_images, train_labels, epochs=8, batch_size=64)
```

```python
model.evaluate(test_images, test_labels)
```
